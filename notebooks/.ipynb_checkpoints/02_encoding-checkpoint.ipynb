```python
def duplicate_columns(frame):
    groups = frame.columns.to_series().groupby(frame.dtypes).groups
    dups = []

    for t, v in groups.items():

        cs = frame[v].columns
        vs = frame[v]
        lcs = len(cs)

        for i in range(lcs):
            ia = vs.iloc[:,i].values
            for j in range(i+1, lcs):
                ja = vs.iloc[:,j].values
                if np.array_equal(ia, ja):
                    dups.append(cs[i])
                    break
    return dups
```

```python
def evaluate_model(ytest, ypred, ypred_proba = None):
    if ypred_proba is not None:
        print('ROC-AUC score of the model: {}'.format(roc_auc_score(ytest, ypred_proba[:, 1])))
    print('')
    print('Accuracy of the model: {}\n'.format(accuracy_score(ytest, ypred)))
    
    score = f1_score(ytest, ypred)
    print('F-Score: %.5f' % score)
    print('')
    print('Classification report: \n{}\n'.format(classification_report(ytest, ypred)))
    #print('Confusion matrix: \n{}\n'.format(confusion_matrix(ytest, ypred)))
    
    
    print('Confusion matrix:')
    cm_gbt = confusion_matrix(ytest, ypred)

    plt.figure(figsize=(9,9))
    sns.heatmap(cm_gbt, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues_r');
    plt.ylabel('Actual label')
    plt.show();
```

```python
classifiers = [
    LogisticRegression(n_jobs=-1),
    DecisionTreeClassifier(),
    RandomForestClassifier(n_jobs=-1, random_state=158),
    AdaBoostClassifier(random_state=158),
    GradientBoostingClassifier(random_state=158),
    #XGBClassifier(),
    LGBMClassifier(random_state=158, n_jobs=-1)
    ]

for classifier in classifiers:
       
    pipe = Pipeline(steps=[('classifier', classifier)])
    pipe.fit(xtrain, ytrain)   
    
    print('The model presented is:', classifier)
    print('')
    print("model score: %.3f" % pipe.score(xtest, ytest))
    
    ypred = pipe.predict(xtest)
    ypred_proba = pipe.predict_proba(xtest)      
    
    print('')
    evaluate_model(ytest=ytest, ypred=ypred, ypred_proba = ypred_proba)
    print('')
        
    yhat = ypred_proba[:, 1]
    # calculate roc curves
    fpr, tpr, thresholds = roc_curve(ytest, yhat)
    # plot the roc curve for the model
    print('ROC Curve:')
    plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
    plt.plot(fpr, tpr, marker='.', label='Random Forest')
    # axis labels
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    # show the plot
    plt.show();
    
    filename  = '../models/' + re.sub(r"\([^()]*\)", '', str(classifier)) + '.pickle'
    pickle.dump(pipe['classifier'], open(filename, 'wb'))
    
    print('-----------------------------------------------------------')
```

```python
# transform the dataset
oversample = SMOTE(random_state=1238, n_jobs=-1, sampling_strategy=0.2)
xtrain2, ytrain2 = oversample.fit_resample(xtrain, ytrain)
```

```python
ytrain2.value_counts()/len(ytrain2)
```

```python
models = []

models.append(("XGBClassifier7", XGBClassifier(scale_pos_weight=7, eval_metric="aucpr", max_depth=10, max_delta_step=3)))
models.append(("XGBClassifier8", XGBClassifier(scale_pos_weight=8, eval_metric="aucpr", max_depth=10, max_delta_step=10)))

results = []
names = []
for name,model in models:
    result = model.fit(xtrain2, ytrain2)
    names.append(name)
    results.append(result)
    
for i in range(len(names)):

    print('')
    print('The model presented is:', names[i])
    ypred = results[i].predict(xtest)
    
    
    print('')

    # keep probabilities for the positive outcome only
    if names[i] not in ['SVC']:
        ypred_proba = results[i].predict_proba(xtest)
        yhat = ypred_proba[:, 1]
        print('')
        evaluate_model(ytest=ytest, ypred=ypred, ypred_proba = ypred_proba)
        
        print('')
        # calculate roc curves
        fpr, tpr, thresholds = roc_curve(ytest, yhat)
        # plot the roc curve for the model
        print('ROC Curve:')
        plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
        plt.plot(fpr, tpr, marker='.', label='Random Forest')
        # axis labels
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        # calculate the g-mean for each threshold
        gmeans = np.sqrt(tpr * (1-fpr))
        # locate the index of the largest g-mean
        ix = np.argmax(gmeans)
        print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))
        plt.scatter(fpr[ix], tpr[ix], s=100, marker='o', color='black', label='Best')
        plt.legend()
        # show the plot
        plt.show();
    
    print('-----------------------------------------------------------')
```
